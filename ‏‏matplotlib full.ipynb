# Introduction to Matplotlib

**Matplotlib** is one of the most important and widely used libraries in Python for creating graphs and data visualizations. 

It provides a variety of tools and options for creating different types of plots, from simple line plots to complex, customized visualizations.

In this lesson, we will focus on **matplotlib.pyplot**, a sub-library of Matplotlib that allows for simple and intuitive plot creation.  

This library is crucial for future work in AI and data science, as visualizing data helps in understanding patterns and trends.

Let's start by exploring some of the fundamental features of `matplotlib.pyplot`.

## Importing the Library

Before we begin, we need to import the `matplotlib.pyplot` library. It's a common practice to import it with the alias `plt`.


```python
import matplotlib.pyplot as plt
```

## Basic Plotting

Let's start with a simple line plot. This is the most basic form of plotting and is used to represent data points on a 2D plane.

Show the relationship between X and X^2, add labels:

```python
x = [i for i in range(1, 11)]
y = [i**2 for i in x]
plt.plot(x, y, label='Relationship between x and x^2', c='pink', linestyle='--')
plt.legend()
plt.title('Graph')
plt.xlabel('x')
plt.ylabel('X^2')
```

## Scatter Plot

A scatter plot displays individual data points. It's useful when you want to show the relationship between two variables.

We will solve the exercise from the previous section but with scatter

```python
plt.scatter(x, y, c='lightblue', marker='*')
```

## Bar Plot

Bar plots are great for comparing quantities corresponding to different groups or categories.

Make it for the categories: `A, B, C, D` and different values

```python
data = {
    'a': 1,
    'b': 7,
    'c':3,
    'd':4
}
plt.bar(data.keys(), data.values(), color='green')
```

## Histogram

## Customizing Plots

Matplotlib allows us to customize plots extensively. We can change line styles, colors, add grids, and much more.

Use `x, y, c, linestyle, marker` and `plt.grid(True)`

```python
plt.plot(x, y, marker='*', linestyle='--', c='red', mec='black', mfc='yellow', label='Something')
plt.legend()
```
